```python
df = pd.read_csv("train.csv")

# retirer les colonnes Cabine, Ticket, PassengerId
# Sex et Embarked en valeur num
df.drop(columns=["Cabin", "Ticket", "PassengerId", "Name"], inplace=True)
df["Sex"] = df.Sex.map({"female": 1, "male": 0})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})
df = df.dropna()
df.count()
```

```python
data_train = df.iloc[:500] # je rÃ©cupÃ¨re les 500 premiers pour entrainer mon algo d'apprentissafge dessus
data_test = df.iloc[500:] # Je rÃ©cupÃ¨re les suivants pour vÃ©rifier s'il a Ã©tÃ© efficace

features_train = data_train.drop(columns=["Survived"])
result_train = data_train["Survived"]

features_test = data_test.drop(columns=["Survived"])
result_test = data_test["Survived"]
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler

cls = SVC()

scaler = MinMaxScaler()
scaler.fit(features_train.values)
features_train_scaled = scaler.transform(features_train)
cls.fit(features_train_scaled, result_train)
```

```python
cls.predict(features_test[:10])
```

```python
result_test[:10]
```

```python
cls.score(scaler.transform(features_test), result_test)
```
